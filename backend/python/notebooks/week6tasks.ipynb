````python
import sys
import os
from dotenv import load_dotenv

load_dotenv()

# One level up from notebooks/ brings us to python/ where products/ lives
sys.path.append(os.path.abspath(".."))

from products.category_model import ProductCategory

# Fetch all categories from MongoDB
categories = ProductCategory.objects.all()
category_map = {cat.title: cat for cat in categories}

print("📦 Available Categories:")
for title in category_map:
    print(f"  - {title}")
````

````python
import json

category_names = list(category_map.keys())

prompt = f"""
Generate exactly 50 realistic warehouse products as a JSON array.
Each product must belong to one of these categories only: {category_names}

Each product must have exactly these fields:
- name: string (product name, be specific e.g. "Samsung 4K TV 55 inch")
- description: string (1-2 sentence product description)
- category: string (must be exactly one of: {category_names})
- price: float (realistic price with 2 decimal places, min 0.01)
- brand: string (realistic brand name)
- quantity_in_warehouse: integer (between 0 and 500)

Rules:
- Generate 10 products per category
- Return ONLY a valid JSON array, no explanation, no markdown
- No extra text before or after the JSON

Example format:
[
  {{
    "name": "Samsung 4K TV 55 inch",
    "description": "A stunning 4K TV with HDR support.",
    "category": "Electronics",
    "price": 599.99,
    "brand": "Samsung",
    "quantity_in_warehouse": 45
  }}
]
"""

response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[{"role": "user", "content": prompt}],
    temperature=0.7
)

raw_output = response.choices[0].message.content
print("🤖 Raw AI Response (first 500 chars):")
print(raw_output[:500])
````

````python
# Clean the response in case AI adds markdown code blocks
cleaned = raw_output.strip()
if cleaned.startswith("```json"):
    cleaned = cleaned[7:]
if cleaned.startswith("```"):
    cleaned = cleaned[3:]
if cleaned.endswith("```"):
    cleaned = cleaned[:-3]
cleaned = cleaned.strip()

# Parse JSON
products_data = json.loads(cleaned)

print(f"✅ Successfully parsed {len(products_data)} products!")
print("\n📋 First product preview:")
print(json.dumps(products_data[0], indent=2))
````

````python
from pydantic import BaseModel, field_validator
from typing import Optional

class ProductSchema(BaseModel):
    name: str
    description: Optional[str] = ""
    category: str
    price: float
    brand: str
    quantity_in_warehouse: int

    @field_validator('price')
    def price_must_be_positive(cls, v):
        if v <= 0:
            raise ValueError("Price must be greater than 0")
        return round(v, 2)

    @field_validator('quantity_in_warehouse')
    def quantity_must_be_non_negative(cls, v):
        if v < 0:
            raise ValueError("Quantity cannot be negative")
        return v

    @field_validator('category')
    def category_must_exist(cls, v):
        valid_categories = ["Electronics", "Food", "Kitchen Essentials", "Clothing", "Books"]
        if v not in valid_categories:
            raise ValueError(f"Category '{v}' is not valid. Must be one of {valid_categories}")
        return v

print("✅ Pydantic schema defined!")
````

````python
valid_products = []
invalid_products = []

for i, product in enumerate(products_data):
    try:
        validated = ProductSchema(**product)
        valid_products.append(validated)
    except Exception as e:
        invalid_products.append({"index": i, "product": product, "error": str(e)})

print(f"✅ Valid products   : {len(valid_products)}")
print(f"❌ Invalid products : {len(invalid_products)}")

if invalid_products:
    print("\n⚠️ Invalid Products Details:")
    for p in invalid_products:
        print(f"  Index {p['index']}: {p['error']}")
````

````python
from products.models import Product
from datetime import datetime, timezone

saved = 0
failed = 0

for validated_product in valid_products:
    try:
        # Match category title to actual MongoDB category object
        category_obj = category_map.get(validated_product.category)

        if not category_obj:
            print(f"⚠️ Category not found for: {validated_product.category}")
            failed += 1
            continue

        product = Product(
            name=validated_product.name,
            description=validated_product.description,
            category=category_obj,
            price=validated_product.price,
            brand=validated_product.brand,
            quantity_in_warehouse=validated_product.quantity_in_warehouse,
            created_at=datetime.now(timezone.utc),
            updated_at=datetime.now(timezone.utc)
        )
        product.save()
        saved += 1

    except Exception as e:
        print(f"❌ Failed to save {validated_product.name}: {e}")
        failed += 1

print(f"\n🎉 Successfully saved : {saved} products")
print(f"❌ Failed to save     : {failed} products")
````

````python
total_products = Product.objects.count()
print(f"📦 Total products in MongoDB: {total_products}")

print("\n🔍 Last 5 saved products:")
for p in Product.objects.order_by('-created_at')[:5]:
    print(f"  - {p.name} | {p.brand} | ${p.price} | qty: {p.quantity_in_warehouse}")
````

````python
# Get all product names from MongoDB to use in the prompt
all_products = Product.objects.all()
product_names = [p.name for p in all_products][:20]  # use first 20 as sample

prompt = f"""
Generate exactly 10 realistic warehouse stock events as a JSON array.
These are scheduled future inventory movements for a warehouse.

Use only these real product names from our warehouse:
{product_names}

Each event must have exactly these fields:
- product_name: string (must be exactly one of the product names listed above)
- quantity_change: integer (positive = stock arriving, negative = stock going out)
- reason: string (realistic reason e.g. "Scheduled delivery from supplier", "Bulk order shipped out")
- event_date: string (a future date in format YYYY-MM-DD, between 2026-04-11 and 2026-06-30)

Rules:
- Mix of positive and negative quantity changes
- Return ONLY a valid JSON array, no explanation, no markdown
- No extra text before or after the JSON

Example format:
[
  {{
    "product_name": "Samsung 4K TV 55 inch",
    "quantity_change": 100,
    "reason": "Scheduled delivery from supplier",
    "event_date": "2026-04-15"
  }}
]
"""

response_events = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[{"role": "user", "content": prompt}],
    temperature=0.7
)

raw_events = response_events.choices[0].message.content
print("🤖 Raw Stock Events from Groq:")
print(raw_events[:500])
````

````python
# Clean response
cleaned_events = raw_events.strip()
if cleaned_events.startswith("```json"):
    cleaned_events = cleaned_events[7:]
if cleaned_events.startswith("```"):
    cleaned_events = cleaned_events[3:]
if cleaned_events.endswith("```"):
    cleaned_events = cleaned_events[:-3]
cleaned_events = cleaned_events.strip()

stock_events = json.loads(cleaned_events)

print(f"✅ Successfully parsed {len(stock_events)} stock events!")
print("\n📋 Stock Events Preview:")
for event in stock_events:
    change = f"+{event['quantity_change']}" if event['quantity_change'] > 0 else str(event['quantity_change'])
    print(f"  📦 {event['product_name']}")
    print(f"     Change : {change} units")
    print(f"     Reason : {event['reason']}")
    print(f"     Date   : {event['event_date']}")
    print()
````

````python
from datetime import datetime, timezone
import time

print("🔄 Applying Stock Events & Checking Audit Trail...")
print("=" * 60)

for event in stock_events:
    # Find the product by name
    product = Product.objects(name=event['product_name']).first()

    if not product:
        print(f"⚠️  Product not found: {event['product_name']}")
        continue

    # Capture BEFORE state
    before_quantity = product.quantity_in_warehouse
    before_updated_at = product.updated_at

    # Calculate new quantity (don't go below 0)
    new_quantity = max(0, before_quantity + event['quantity_change'])

    # Apply the update
    product.quantity_in_warehouse = new_quantity
    product.updated_at = datetime.now(timezone.utc)
    product.save()

    # Capture AFTER state
    after_quantity = product.quantity_in_warehouse
    after_updated_at = product.updated_at

    # Print audit trail
    change = f"+{event['quantity_change']}" if event['quantity_change'] > 0 else str(event['quantity_change'])
    print(f"📦 Product       : {product.name}")
    print(f"   Reason        : {event['reason']}")
    print(f"   Event Date    : {event['event_date']}")
    print(f"   Quantity      : {before_quantity} → {after_quantity} ({change})")
    print(f"   updated_at    : {before_updated_at} → {after_updated_at}")
    print(f"   Audit Trail   : {'✅ updated_at changed!' if before_updated_at != after_updated_at else '⚠️ unchanged'}")
    print()

print("=" * 60)
print("🎉 All stock events applied successfully!")
````
